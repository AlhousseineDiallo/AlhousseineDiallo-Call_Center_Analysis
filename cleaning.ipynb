## The next step involve data cleaning and preparation !

```python
# let's rename the column in accordance with the snake case convention

df.columns = df.columns.str.replace(" ", "_").str.lower()
```

```python
df
```

```python
# let's drop the null values and the duplicates
df.isnull().sum()
```

```python
# second method
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(f"{col} - {pct_missing} % of nan")
    
```

```python
df.duplicated().sum()
```

```python
# Convert date column to a datetime type
df['date'] = pd.to_datetime(df['date'])
```

```python
# let's convet all the column from HH:MM:SS as an integer to facilitate our analysis
columns_convert = ['answer_speed_(avg)', 'talk_duration_(avg)', 'waiting_time_(avg)']
df[columns_convert] = df[columns_convert].apply(lambda x: pd.to_timedelta(x).dt.total_seconds().astype('int32'))
```

```python
# let' check it 
df.dtypes
```

```python
df.head()
```

```python
# let's remove the % in the answer_rate and the service_level and then we'll convert it as float
cols = ['answer_rate', 'service_level_(20_seconds)']
for col in cols:
    df[col] = df[col].str.replace('%', '').astype(float)
```

```python
# let's check it
df.dtypes
```

```python
df.head()
```

```python
# let's rename our column according to our changes
df.rename(columns={
    'answer_speed_(avg)': 'answer_speed(seconds)',
    'waiting_time_(avg)': 'waiting_time(seconds)',
    'talk_duration_(avg)': 'talk_duration(seconds)',
    'service_level_(20_seconds)': 'service_level_rate'
}, inplace=True)
```

```python
df.head()
```

```python
# check it
df.info()
```

```python
# let's change the int64 to a int32 for all the numeric columns
df[['incoming_calls', 'answered_calls', 'abandoned_calls']] = df[['incoming_calls', 'answered_calls', 'abandoned_calls']].astype('int32')
```

```python
df.dtypes
```

```python
# let's create a day column
df['day_of_week'] = df['date'].dt.day_name()
```

```python
df.head()
```

```python
# let's change the type of the day_of_week column
df['day_of_week'] = df['day_of_week'].astype('category')
```

```python
# let's see the uniformity of the date column
df[~(df['date'].astype(str).str.contains(r'\d{4}-\d{2}-\d{2}'))]
```
